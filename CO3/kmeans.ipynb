```python
import pandas as pd
```

```python
dst=pd.read_csv('data.csv')
```

```python
dst
```

```python
from sklearn.cluster import KMeans
```

```python
objk=KMeans(n_clusters=2)
```

```python
objk.fit(dst.drop('private',axis=1))
```

```python
objk.cluster_centers_
```

```python
def fun(data):
    if data =='Yes':
        return 1
    else:
        return 0
    
```

```python
dst['cluster']=dst['private'].apply(fun)
```

```python
dst
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(dst['cluster'],objk.labels_))
```

```python
print(classification_report(dst['cluster'],objk.labels_))
```
